```python
#read in csv file from directory
agmp_data_01 = pd.read_csv("/Users/perceval/data_science/agmp/gwas_without_NaN.csv")
```

```python
#display contents of the csv file

##### renaming columns ############

# Pubmed_id = Publication_ID
# phenotype = phenT
agmp_data_01
```

```python
# convert the csv file into a data frame
df = pd.DataFrame(agmp_data_01)
```

```python
# display the contents of the dataframe
df
```

```python
#filter out rs_d and origin of participants
```

```python

# split the column "origin_of_participants" into columns separated by a ,
split_values = df['origin_of_participants'].str.split(',', expand=True)
```

```python
# display the split values
split_values
```

```python

# rename the column names
split_values.columns = [f'country_participant_0{i+1}' for i in range(split_values.shape[1])]
```

```python
split_values.columns
```

```python

# Concatenate the split with the original data frame
result_df = pd.concat([df, split_values], axis=1)
```

```python
result_df.head()
```

```python
result_df.columns
```

```python
import pandas as pd
import numpy as np

# Assuming 'latitude' and 'longitude' are the fields we are generating
# Replace these field names if they differ in your DataFrame
if 'latitude' in df.columns and 'longitude' in df.columns:
    df['latitude'] = np.random.uniform(low=-90.0, high=90.0, size=(len(df)))
    df['longitude'] = np.random.uniform(low=-180.0, high=180.0, size=(len(df)))
else:
    # If columns are not present, add them with random coordinates
    df['latitude'] = np.random.uniform(low=-90.0, high=90.0, size=(len(df)))
    df['longitude'] = np.random.uniform(low=-180.0, high=180.0, size=(len(df)))

# Output to check the top rows to ensure coordinates are added/updated
df.head()
```

```python
result_df.columns
```

```python
df = result_df.iloc[:200000] 
```

```python
df.shape
df
```

```python
geolocator = Nominatim(user_agent="myApp01")

for i in df.index:
    try:
        #tries fetch address from geopy
        location = geolocator.geocode(df['country_participant_01'][i])
        
        #append lat/long to column using dataframe location
        df.loc[i,'latitude_01'] = location.latitude
        df.loc[i,'longitude_01'] = location.longitude

    except:
        #catches exception for the case where no value is returned
        #appends null value to column
        df.loc[i,'latitude_01'] = ""
        df.loc[i,'longitude_01'] = ""
      

 
```
